The dataset used in this project is originally from NIDDK. The objective is to predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset.
Build a model to accurately predict whether the patients in the dataset have diabetes or not.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("health care diabetes.csv")
```

```python
df.info()
```

```python
X = df.drop("Outcome",axis=1)
```

```python
y = df[["Outcome"]]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scale = scaler.fit_transform(X)
```

```python
x_scale_df = pd.DataFrame(x_scale,columns = X.columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scale_df,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN_clf.fit(x_train,y_train)
```

```python
y_pred = KNN_clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
model_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy of model is : ",model_accuracy)
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
print("confusion matrix on testing data is : \n",conf_matrix)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print("classification report is : \n",clf_report)
```

```python
from sklearn.model_selection import GridSearchCV
```
