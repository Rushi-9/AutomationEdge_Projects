```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop('Outcome',axis=1)
```

```python
Y = df[['Outcome']]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scale = scaler.fit_transform(X)
```

```python
x_scale_df = pd.DataFrame(x_scale,columns = X.columns)
```

```python
x_scale_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scale_df,Y,test_size = 0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
svc_clf = SVC(kernel='linear',random_state=0)
```

```python
svc_clf.fit(x_train,y_train)
```

```python
y_pred = svc_clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_percnt = accuracy_score(y_test,y_pred)
```

```python
print(accuracy_percnt)
```
